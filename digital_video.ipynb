## What is an image?



```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.image as mpimg
import numpy as np
from PIL import Image
%matplotlib inline
```

```python
# loading image
img = Image.open('res/trex_small.png')

def plot_img(plt, size, show_grid=False):
    plt.imshow(img.resize((size, size)))
    plt.set_title("{}x{}".format(size, size))
    plt.xaxis.set_major_formatter(ticker.NullFormatter())
    plt.yaxis.set_major_formatter(ticker.NullFormatter())
    if show_grid:
        plt.yaxis.grid(True)
        plt.xaxis.grid(True)
        plt.xaxis.set_ticks(np.arange(1, size, 1) - 0.5)
        plt.yaxis.set_ticks(np.arange(1, size, 1) - 0.5)
    else:
        plt.axis('off')
        
f, (plt1, plt2, plt3, plt4) = plt.subplots(1, 4, figsize=(20, 20))
plot_img(plt1, 128)   
plot_img(plt2, 64) 
plot_img(plt3, 32)
plot_img(plt4, 16)
```

## Image Resolution

The resolution of an image is the number of horizontal pixels by the number of vertical pixels

<font size="5"> Resolution = Width x Height </font>

A resolution of 1920x1080 means our image has a width of 1920 pixels and a height of 1080

It's also common to name the resolution of an image with the total number of pixels, eg: 22.1 MP (Mega Pixels) 
For example a Canon 5D Mk III produces images of 5760 pixels by 3840, which equates to 22,118,400 pixels, or 22.1 MP



```python
f, (plt1, plt2, plt3, plt4) = plt.subplots(1, 4, figsize=(20, 20))
plot_img(plt1, 128, True)   
plot_img(plt2, 64, True) 
plot_img(plt3, 32, True)
plot_img(plt4, 16, True)
```

## Bit-depth: Quantization

```python
import math

def int_to_rgb(color, max_value):
    blue =  color % max_value
    green = (color / max_value) % max_value
    red = (color / max_value / max_value) % max_value
    return (red, green, blue)


def gen_color_map(bit_depth):
    max_value = 2 ** bit_depth
    colors_count = 2 ** (bit_depth * 3)
    size = int(math.sqrt(colors_count))
    colors = np.zeros((size, size, 3))
    i = 0
    for x in range(size):
        for y in range(size):
            colors[x, y] = np.asarray(int_to_rgb(i, max_value)) / max_value
            i += 1
    return colors
            
color_map = gen_color_map(5)
f, ax = plt.subplots(figsize=(20, 20))
ax.matshow(color_map)
```

```python
def to_rgb(pixel):
    return (
        int(round(256 * pixel[0])), # Red
        int(round(256 * pixel[1])), # Green
        int(round(256 * pixel[2]))  # Blue
    )

img = mpimg.imread('res/trex_small.png')

size = 10; x = 10; y = 70;
f, ax = plt.subplots(figsize=(20, 20))
data = img[x:x+size,y:y+size]

for i in range(size):
    for j in range(size):
        c = to_rgb (data[j,i])
        ax.text(i, j, str(c), va='center', ha='center')

ax.matshow(data)
```
