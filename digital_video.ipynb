# Pixels: Spacial sampling

Whereas audio is continous in time, an image is continous in space.

There are an infinite numbers of pixels, the first step to convert our image into a digital image is sampling.

We will split our image in pixels, each pixels will have as value a color.

We are converting our signal from a continous-space to a discrete-space.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.image as mpimg
import numpy as np
from PIL import Image
%matplotlib inline
```

```python
# loading image
img = Image.open('res/trex_small.png')

def plot_img(plt, size, show_grid=False):
    plt.imshow(img.resize((size, size)))
    plt.set_title("{}x{}".format(size, size))
    plt.xaxis.set_major_formatter(ticker.NullFormatter())
    plt.yaxis.set_major_formatter(ticker.NullFormatter())
    if show_grid:
        plt.yaxis.grid(True)
        plt.xaxis.grid(True)
        plt.xaxis.set_ticks(np.arange(1, size, 1) - 0.5)
        plt.yaxis.set_ticks(np.arange(1, size, 1) - 0.5)
    else:
        plt.axis('off')

        
f, (plt1, plt2, plt3, plt4) = plt.subplots(1, 4, figsize=(20, 20))
plot_img(plt1, 128)   
plot_img(plt2, 64) 
plot_img(plt3, 32)
plot_img(plt4, 16)
```

## Bit-depth: Quantization

```python
def int_to_rgb(color):
    return ((color & 255),
            (color >> 8) & 255,
            (color >> 16) & 255)

def gen_color_map(bit_depth):
    colors_count = 2 ** bit_depth
    size = int(colors_count / 2 / 2)
    colors = np.zeros((size, size, 3))
    i = 0
    for x in range(size):
        for y in range(size):
            print (int_to_rgb(i))
            colors[x, y] = np.asarray(int_to_rgb(i)) / 15.
            i += 1
    return colors
            
color_map = gen_color_map(4)
f, ax = plt.subplots(figsize=(20, 20))
ax.matshow(color_map)
```

```python
def to_rgb(pixel):
    return (
        int(round(256 * pixel[0])), # Red
        int(round(256 * pixel[1])), # Green
        int(round(256 * pixel[2]))  # Blue
    )

img = mpimg.imread('res/trex_small.png')

size = 10; x = 10; y = 70;
f, ax = plt.subplots(figsize=(20, 20))
data = img[x:x+size,y:y+size]

for i in range(size):
    for j in range(size):
        c = to_rgb (data[j,i])
        ax.text(i, j, str(c), va='center', ha='center')

ax.matshow(data)
```
